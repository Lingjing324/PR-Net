```python
from scipy.io import loadmat
import glob
import matplotlib.pylab as plt

def vis_bat(xx,yy,yyp):
    fig = plt.figure(1, figsize=(20, 40))
    for i in range(10):
        x=xx[i]
        y=yy[i]
        x=x.T
        y=y.T
        yp=yyp[i]
        yp=yp.T
        ax = fig.add_subplot(10, 2, i*2+1)
        plt.scatter(x[:,0],x[:,1], label="source",s=20,c="b")
        plt.scatter(y[:,0],y[:,1], label="target",s=40,c="r",marker='x')
        plt.ylim(-2.5, 2.5)
        plt.xlim(-2.5, 2.5)
        ax.axis('off')
        ax = fig.add_subplot(10, 2, i*2+2)
        plt.scatter(y[:,0],y[:,1], label="target",s=40, c="r",marker="x")
        plt.scatter(yp[:,0],yp[:,1], label="transformed",s=20,c="b")
        plt.ylim(-2.5, 2.5)
        plt.xlim(-2.5, 2.5)
        ax.axis('off')
    plt.show()
    #plt.savefig(name,transparent=True)
    plt.close('all')
```

```python
for m in sorted(glob.glob("./*mat"))[::-1]:
    print("******************************************")
    print(m)
    s=loadmat(m)['s']
    t=loadmat(m)['t']
    ts=loadmat(m)['ts']
    vis_bat(t,s,ts)
```
